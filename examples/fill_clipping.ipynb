```julia
using GreinerHormann
using PyPlot

circle = Polygon()

x_max = 10
y_max = 10.2
circle_seg = 2*pi/100

r = min(x_max, y_max)/2

for i = 0:circle_seg:2*pi-circle_seg
    push!(circle, Vertex([r*cos(i)+x_max/2, r*sin(i)+y_max/2]))
end

x = Float64[]
y = Float64[]
for vert in circle
    push!(x, vert.location[1])
    push!(y, vert.location[2])
end
plot(x, y, color="red", linewidth=1.0, linestyle="-")

fill = Polygon()
push!(fill, Vertex([-2.0,-2.0]))
scan = [-1.0,11.0]
for i = 0.5:9.5
    for j in scan
        push!(fill, Vertex([i,j]))
    end
    push!(fill, Vertex([i+1,scan[2]]))
    reverse!(scan)
end

x = Float64[]
y = Float64[]
for vert in fill
    push!(x, vert.location[1])
    push!(y, vert.location[2])
end
plot(x, y, color="blue", linewidth=1.0, linestyle="-")


clip = intersection(circle, fill)
x = Float64[]
y = Float64[]
for poly in clip
    for vert in poly
        x1, y1 = vert.location
        x2, y2 = vert.next.location
        plot([x1,x2], [y1,y2], color="green", linewidth=2.0, linestyle="-")
    end
end
```
